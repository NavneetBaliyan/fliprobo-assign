```python
import string
from collections import Counter
import matplotlib.pyplot as plt
```

```python
def extract_tweets():
    import GetOldTweets3 as got
    location="India"
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch('corona').setSince("2020-02-11").setUntil("2020-03-11").setMaxTweets(1000)
    
    tweets= got.manager.TweetManager.getTweets(tweetCriteria)
    
    text_tweets=[[tweet.text] for tweet in tweets]
    return text_tweets
    
```

```python
extract_tweets()
```

```python
text = ""
text_tweets = extract_tweets()

length = len(text_tweets)

for i in range(0, length):
    text = text_tweets[i][0] + " " + text

lower_case = text.lower()

cleaned_text = lower_case.translate(str.maketrans('', '', string.punctuation))

tokenized_words = cleaned_text.split()
```

```python
text_tweets
```

```python
len(text_tweets)
```

```python
from nltk.corpus import stopwords
stop=stopwords.words("english")
```

```python
final_words = [word for word in tokenized_words if word not in stop]
```

```python
final_words
```

```python
# Get emotions text
emotion_list = []
with open('emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace('\n', '').replace(',', '').replace("'", '').strip()
        word, emotion = clear_line.split(':')
        if word in final_words:
            emotion_list.append(emotion)

w = Counter(emotion_list)
print(w)

```

```python
fig, ax1 = plt.subplots()
ax1.bar(w.keys(), w.values())
fig.autofmt_xdate()
plt.savefig('graph.png')
plt.show()
```
