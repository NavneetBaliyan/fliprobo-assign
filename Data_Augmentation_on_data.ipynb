```python
def increase_brightness(image):
    image=cv2.imread(images[x])
    M=np.ones(image.shape,dtype='uint8')*70
    
    added=cv2.add(image,M)
    return added
```

```python
def decrease_brightness(image):
    image=cv2.imread(images[x])
    M=np.ones(image.shape,dtype='uint8')*90
    
    subtract=cv2.subtract(image,M)
    return subtract
```

```python
def flipped(image):
    image=cv2.imread(images[x])
    flipped=cv2.flip(image,2)
    return flipped
```

```python
def sharpening(image):
    image=cv2.imread(images[x])
    
    kernel_sharpening=np.array([[-1,-1,-1],
                                [-1,10,-1],
                                [-1,-1,-1]])
#applying different kernel to input image
    sharpened=cv2.filter2D(image,-1,kernel_sharpening)

    return sharpened
```
